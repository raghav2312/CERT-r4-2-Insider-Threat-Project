## Data cleaning

### Subtask:
Clean the data by handling outliers and missing values in the dataset.


```python
import numpy as np

# Calculate the IQR for 'duration' and 'n_allact'
Q1_duration = df['duration'].quantile(0.25)
Q3_duration = df['duration'].quantile(0.75)
IQR_duration = Q3_duration - Q1_duration

Q1_n_allact = df['n_allact'].quantile(0.25)
Q3_n_allact = df['n_allact'].quantile(0.75)
IQR_n_allact = Q3_n_allact - Q1_n_allact

# Define bounds for outliers
lower_bound_duration = Q1_duration - 1.5 * IQR_duration
upper_bound_duration = Q3_duration + 1.5 * IQR_duration

lower_bound_n_allact = Q1_n_allact - 1.5 * IQR_n_allact
upper_bound_n_allact = Q3_n_allact + 1.5 * IQR_n_allact

# Replace outliers with the upper/lower bound
df['duration'] = np.where(df['duration'] < lower_bound_duration, lower_bound_duration, df['duration'])
df['duration'] = np.where(df['duration'] > upper_bound_duration, upper_bound_duration, df['duration'])

df['n_allact'] = np.where(df['n_allact'] < lower_bound_n_allact, lower_bound_n_allact, df['n_allact'])
df['n_allact'] = np.where(df['n_allact'] > upper_bound_n_allact, upper_bound_n_allact, df['n_allact'])

# Check for missing values
print("Number of missing values per column:")
print(df.isnull().sum())

# Handle missing values (e.g., fill with mean)
df.fillna(df.mean(), inplace=True)

# Verify that missing values are handled
print("\nNumber of missing values per column after handling:")
print(df.isnull().sum())
```

## Feature engineering

### Subtask:
Engineer new features that might be helpful for anomaly detection.


```python
# Calculate quantiles for duration and create a new feature
duration_quantiles = df['duration'].quantile([0.25, 0.50, 0.75, 0.90, 0.95, 0.99])
df['duration_quantile'] = pd.qcut(df['duration'], q=[0, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99, 1], labels=False, duplicates='drop')

# Create user activity ratios
df['http_ratio'] = df['n_http'] / df['n_allact'].replace(0, 1)
df['email_ratio'] = df['n_email'] / df['n_allact'].replace(0, 1)
df['file_ratio'] = df['n_file'] / df['n_allact'].replace(0, 1)

# Create security event ratios
df['http_leakf_ratio'] = df['http_n_leakf'] / df['n_http'].replace(0, 1)
df['http_hackf_ratio'] = df['http_n_hackf'] / df['n_http'].replace(0, 1)

# Display the first few rows of the DataFrame with the new features
display(df[['duration', 'duration_quantile', 'http_ratio', 'email_ratio', 'file_ratio', 'http_leakf_ratio', 'http_hackf_ratio']].head())
```
